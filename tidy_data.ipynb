- Cake Recipes

     - Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

    - Tidy the data as necessary.

    - Which recipe, on average, is the best? recipe b

    - Which oven temperature, on average, produces the best results? 275

    - Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

#### Importing the necessary csv file

```python
cake = pd.read_csv('/Users/bibekmainali/codeup-data-science/classification-exercises/untidy-data/cake_recipes.csv')
```

```python
cake
```

- Right away, we can see that are multiple value in one of the column

```python
# splitting the columns  at ':', and expanding it

splitted_df = cake['recipe:position'].str.split(':', expand = True)
splitted_df.columns = ['recipe', 'position']
```

```python
splitted_df
```

```python
cake = cake.drop(columns = 'recipe:position') # dropping the column that we had just splitted
```

```python
cake = pd.concat([cake, splitted_df], axis = 1) # adding together the splitted df and the original df
```

```python
# positioning our columns where we want them

cake = cake[['recipe', 'position', '225', '250', '275', '300']]
cake


# using df.melt to take the df from wide to long format and also giving proper names for the columns
x = cake.melt(id_vars = ['recipe', 'position'], var_name = "Temp", value_name = 'Score')
```

```python
x
```

#### Which recipe, on average, is the best?
    - we will answer this question by grouping the above dataframe by recipe and calculate the mean

```python
x.groupby('recipe').Score.mean() # recipe b has the highest score, so, its the best
```

#### Which oven temperature, on average, produces the best results? 
    - we are grouping the above data by temp and getting the average

```python
x.groupby('Temp').mean() # the average for 275 is higher than the rest, so at this temp, best results are produced
```

```python
x.sort_values(by = ['Score', 'Temp'], ascending = False).head(5) # getting the combination of the temp and the best score
```
