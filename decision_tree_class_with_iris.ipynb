```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import prepare

from pydataset import data

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

## Decision Tree
    - Are supervised machine learning process
    - Use training data to train the tree to find a decision boundary to use as a decision rule for future data
    - Given enough depth, they are overfitting machines
 - a sequence of rules that can be used to classify 2 or more classes
 - each node represents a single input 

```python
iris = prepare.prep_iris()
```

```python
iris
```

```python
train_validate, test = train_test_split(iris, random_state = 123, stratify = iris.species, test_size = 0.1)
```

```python
train, validate = train_test_split(train_validate, random_state = 123, stratify = train_validate.species, test_size = 0.1)
```

```python
y_train = train.species
x_train = train.drop(columns = ['species', 'species_versicolor', 'species_virginica'])
```

```python
clf = DecisionTreeClassifier(max_depth=3, random_state=125)
```

```python
clf = clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_train)
```

```python
clf.score(x_train, y_train)
```

```python
print(classification_report(y_train, y_pred))
```

```python

import graphviz

from graphviz import Graph
data = export_graphviz(clf, out_file=None) 
graph = graphviz.Source(data) 

graph.render('iris_decision_tree', view=True)
```
