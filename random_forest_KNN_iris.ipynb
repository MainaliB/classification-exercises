## Random Forest is:
- Forest of Decision Tress

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import prepare

from pydataset import data

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
df = prepare.prep_iris()
```

```python
df
```

```python
df = df.drop(columns = ['species_versicolor', 'species_virginica'])
```

```python
train_validate, test = train_test_split(df, test_size = 0.10, random_state = 9, stratify = df['species'])
train, validate = train_test_split(train_validate, test_size = 0.10, random_state = 9, stratify = train_validate.species)
```

## Lets create model 1 with random forest

```python
y_train = train.species
x_train1 = train.drop(columns = 'species')
x_train1
```

```python
rf1 = RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=3, max_features='auto',
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=100,
                       n_jobs=None, oob_score=False, random_state=123,
                       verbose=0, warm_start=False)
```

```python
rf1.fit(x_train1, y_train)
```

```python
y_pred1 = rf1.predict(x_train1)
```

```python
print(classification_report(y_train, y_pred1))
```

```python
rf1.score(x_train1, y_train)
```

Arguments:
 - min_sample_split = number of samples needed in the node for it to stop splitting
 - max_depth = the maxium depth of a tree, if its none, it will continue to split until all the leaves are pure

## K-Nearest Neighbors
 - Classifier
 - Supervised
 - Not K-means
 - k = # of closest neighbors to find
 

```python
from sklearn.neighbors import KNeighborsClassifier
```

## Model 1 with KNN, with default for number of neighbors

```python
# create model 1 with KNN

y_train = train.species
x_train1 = train.drop(columns = 'species')
```

```python
knn1 = KNeighborsClassifier()
```

```python
knn1.fit(x_train1, y_train)
```

```python
y_pred1 = knn1.predict(x_train1)
```

```python
# Accuracy Score

knn1.score(x_train1, y_train)
```

```python
print(classification_report(y_train, y_pred1))
```

## Model 2 with # of neighbors = 7

```python
x_train2 = train.drop(columns = 'species')
y_train = train.species


knn2 = KNeighborsClassifier(n_neighbors = 7)
```

```python
knn2.fit(x_train2, y_train)
```

```python
# Calculating Accuracy
knn2.score(x_train2, y_train)
```

```python
y_pred2 = knn2.predict(x_train2)
```

```python
print(classification_report(y_train, y_pred2))
```
